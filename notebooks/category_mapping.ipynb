```python
import pandas as pd

# -----------------------------
# Step 1: Load the original arxiv.csv
# -----------------------------
cat_df = pd.read_csv("arxiv.csv")  # columns: code, name, description

# -----------------------------
# Step 2: Define main categories based on prefix
# -----------------------------
prefix_to_main = {
    'cs': 'Computer Science',
    'math': 'Mathematics',
    'physics': 'Physics',
    'astro-ph': 'Physics',
    'cond-mat': 'Physics',
    'gr-qc': 'Physics',
    'hep-ex': 'Physics',
    'hep-lat': 'Physics',
    'hep-ph': 'Physics',
    'hep-th': 'Physics',
    'math-ph': 'Physics',
    'nlin': 'Physics',
    'nucl-ex': 'Physics',
    'nucl-th': 'Physics',
    'quant-ph': 'Physics',
    'q-bio': 'Quantitative Biology',
    'q-fin': 'Quantitative Finance',
    'stat': 'Statistics',
    'eess': 'EE & Systems Science',
    'econ': 'Economics',
}

# Add a main_category column to the CSV
cat_df['main_category'] = cat_df['code'].apply(lambda x: prefix_to_main.get(x.split('.')[0], 'Other'))

# -----------------------------
# Step 3: Save the updated CSV
# -----------------------------
cat_df.to_csv("arxiv_v2.csv", index=False)

# Quick check
print(cat_df[['code', 'name', 'main_category']].head(50))
```
